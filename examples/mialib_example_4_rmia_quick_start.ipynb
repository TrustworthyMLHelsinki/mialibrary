## Quick Start

### With Command Line

If you want to run in the command line, try:
```
python -m  mialibrary.rmia.process_rmia --data_path /path/to/data --offline
```
Ensure you have the required statistics and in_indices files in the `data_path`, starting with `stat_*`, `stat_pop_*`, and `in_indices_*`.
